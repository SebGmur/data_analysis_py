<h3>Putting everything together</h3>
And now, we've finally arrived to the final, desired version of the <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">DataFrame</b> parsed from our CSV file. The steps were:

```python
df = pd.read_csv('data/btc-market-price.csv', header=None)
df.columns = ['Timestamp', 'Price']
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df.set_index('Timestamp', inplace=True)
```

<h2>Dropping null values on DataFrames</h2>
You saw how simple it is to drop <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">na</b>s with a Series. But with <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">DataFrame</b>s, there will be a few more things to consider, because you can't drop single values. You can only drop entire columns or rows. Let's start with a sample <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">DataFrame</b><b>:</b>

```python
df = pd.DataFrame({
    'Column A': [1, np.nan, 30, np.nan],
    'Column B': [2, 8, 31, np.nan],
    'Column C': [np.nan, 9, 32, 100],
    'Column D': [5, 8, 34, 110],
})
```

```python
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

The default <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">dropna</b> behaviour will drop all the rows in which any null value is present:

```python
df.dropna() # This drops all the rows where any data series contains na values
```

In this case we're dropping <b>rows</b>. Rows containing null values are dropped from the DF. You can also use the <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">axis/b> parameter to drop columns containing null values:

```python
df.dropna(axis=1) # axis='columns' also works
```

In this case, any row or coulmn that contains <b>at least</b> one null value will be dropped. Which can be, depending on the case, too extreme. You can control this behaviour with the <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">how</b> parameter. Can be either <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">'any'</b> or <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">'all'</b>:

```python
df2 = pd.DataFrame({
    'Column A': [1, np.nan, 30],
    'Column B': [2, np.nan, 31],
    'Column C': [np.nan, np.nan, 100],
})
```

```python
df2
```

```python
df.dropna(how='all') # drop rows where all the value are null or  column if "axis=1"
```

```python
df.dropna(how='any') # drop rows where any value is na (default behaviour) or  column if "axis=1"
```

You can also use the <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">thresh</b> parameter to indicate a threshold ( a minimum number) of non-null values for the row/column to be kept:

```python
df
```

```python
df.dropna(thresh=3)
```

```python
df.dropna(thresh=3, axis='columns')
```

Sometimes a <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">na</b>s with a Series. You can only drop entire columns or rows. Let's start with a sample <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">nan</b> can be replace with a <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">0</b>, sometimes it can be replaced with the <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">mean</b> of the sample, and sometimes you can take the closest value. Again, it depends on the context. We'll show you the different methods and mechanisms and you can then apply them to your own problem.

<br>
After dealing with many datasets I can tell you that "missing data" is not such a big deal. The best thing that can happen is to clearly see values like <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">np.nan</b>. The only thing you need to do is just use method like <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">isnull</b> and <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">fillna/dropna</b> and pandas will take care of the rest. <br>
But sometimes, you can have invalid values that are not just "missing data" (<b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">None</b>, or <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">nan</b>). For example:

```python
df = pd.DataFrame({
    'Sex': ['M', 'F','F', 'D', '?'],
    'Age': [29, 30, 24, 290, 25]
})
```

The previous <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">DataFrame</b> doesn't have any "missing value", but clearly has invalid data. <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">290</b> doesn't seem like a valid age, and <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">D</b> and <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">?</b> don't correspond with any known sex categoy. How can you clean these not-missing, but clearly invalid values then?

<h3>Finding Uniques Values</h3><br>
The first step to clean invalid values is to <b>notice</b> them, then <b>identify</b> them and finally handle them appropriately (remove them, replace them, etc). Usually, for a "categorical" type of field (like Sex, which only takes of a discrete set <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">('M', 'F')</b>), we start by analyzing the varety of values present. For that, we use the <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">unique()</b> method:

```python
df['Sex'].unique()
```

```python
df['Sex'].value_counts()
```

Clearly if you see values like <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">'D'</b> or <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">'?'</b>, it'll immediately raise your attention. NOw, What to do with them? Let's say you picked up the phone, called the survey company and they told you that <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">'D'</b> was a typo and it should actually be <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">'F'</b>. You can use <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">replace</b> function to replace these values:

```python
df['Sex'].replace('D', 'F')
```

It can accept a dictionary of values to replace. <h2>Duplicates</h2><br>
Checking duplicate values is extremely simple. It'll behave differently between Series and DataFrames. Let's start with Series. As an example, let's say we're throwing a fancy party and we're inviting Ambassadors from Europe. But can only invite one ambasador per country. This is our original list, and as you can see, both the UK and Germany have duplicated ambassadors:

```python
ambassadors = pd.Series([
    'France',
    'United Kingdom',
    'United Kingdom',
    'Italy',
    'Germany',
    'Germany',
    'Germany',
], index=[
    'Gerard Araud',
    'Kim Darroch',
    'Peter Westmacott',
    'Armando Varricchio',
    'Peter Witting',
    'Peter Ammon',
    'Klaus Scharioth',
])
```

```python
ambassadors
```

The two most important methods to deal with duplicates are <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">duplicated</b> (that will tell you which values are duplicated) and <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">drop_duplicates</b> (which will just get rid of duplicates):

```python
ambassadors.duplicated() #.duplicated() return true for the second, and next values. Does't show true for the first value
```

In this case <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">duplicated</b> didn't consider <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">'Kim Darroch'</b>, the first instance of the United Kingdom or <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">'Peter Wittig'</b> as duplicates. That's because, by default, it'll consider the first occurence of the value as not-duplicate. You can change this behaviour with the <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">keep</b> parameter:

```python
ambassadors.duplicated(keep='last')
```

In this case, the result is "flipped", <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">'Kim Darroch'</b> and <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">'Peter Witting'</b> (the first ambasadors of their countries) are considered duplicates, but <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">'Peter Westmacott'</b> and <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">'Klaus Scharioth'</b> are not duplicates. You can also choose to mark all of them as duplicates with <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">keep=False</b><b>:</b> 

```python
ambassadors.duplicated(keep=False) # This method tells you which value is duplicated
```

A similar method is <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">'drop_duplicates'</b>, which just excludes the duplicated values and aslo accepts the <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">keep</b> parameter:

```python
ambassadors.drop_duplicates() # This will only keep the first occurance of an duplicated value
```

```python
ambassadors.drop_duplicates(keep='last') # This will only keep the last occurance of an duplicated value
```

```python
ambassadors.drop_duplicates(keep=False) # This won't keep any occurance of a duplicated value
```

You know that the single columns represent the values "Year, Sex, Country and number of children", but it's all been grouped in the same column and separated by an underscore. Pandas has a convenient method named <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">split</b> that we can use in thse situations:

```python
df
```

```python
df['Year'].str.contains('\?') # ? is a special character, so in order to search for that special character we need to escape it
```

<b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">contains</b> take a regex/pattern as a first value, so we need to escape the <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">?</b> symbol as it has a special meaning for these patterns. Regular letters don't need escaping:

```python
df['Country'].str.contains('U')
```

Removing blank spaces (like in <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">'US    '</b>) or <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">'I      T'</b> can be achived with <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">strip</b> (<b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">lstrip</b> and <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">rstrip</b> also exist) or just <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">replace</b><b>:</b>

```python
df['Country'].str.strip()
```

```python
df['Country'].str.replace(' ', '')
```

As we said, <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">replace</b> and <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">contains</b> take regex patterns, which can make it easier to replace values in bulk:

```python
df['Year'].str.replace(r'(?P<year>\d{4})\?', lambda m: m.group('year'))
```

<h3> Reading data with Python</h3>
As we saw on previous courses we can read data simply using Python.<br>
When you want to work with a file, the first thing to do is to open it. This is done by invoking the <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">open()</b> built-in function. <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">open()</b> has a single required argument that is the path to the file and has a single return, the file object. The <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">with</b> statement automatically takes care of closing the file once it leaves the <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">with</b> block, enve in case of error.

```python
with open('data/btc-market-price.csv', 'r') as fp:
    print(fp)
```

Once the file is open, we can read its content as follows:

```python
with open('./data/btc-market-price.csv', 'r') as fp:
    for index, line in enumerate(fp.readlines()):
        # read just the first 10 lines
        if (index < 10):
                print(index, line)
```

But what happens if the separatr is unknown, like in the file <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">exam_review.csv</b>:

In this case, the separator is not a comma, but the <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">\></b> sign. It's still "CSV", although not technically separated by commas. <br>
<h4>The <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">CSV</b> module</h4>
Python includes the builtin module <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">csv</b> that helps a little bit more with the process of reading CSVs:


```python
import csv
```

```python
with open('data/btc-market-price.csv', 'r') as fp:
    reader = csv.reader(fp)
    for index, (timestamp, price) in enumerate(reader):
        # read just the first 10 lines
        if (index < 10):
            print(f"{timestamp}: ${price}")
```

The <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">csv</b> modules takes care of splitting the file using a given separator (called <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">delimiter</b>) adn creating an iterator for us.

```python
with open('data/exam_review.csv', 'r') as fp:
    reader = csv.reader(fp, delimiter='>') #special delimiter
    next(reader) #skipping header
    for index, values in enumerate(reader):
        if not values:
            continue # skip empty lines
        fname, lname, age, math, french = values
        print(f"{fname} {lname} (age {age}) got {math} in Math and {french} in French")
```

In order to work with a SQLite database from Python, we first have to connect to it. We can do that using the connect function, which returns a <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">Connection</b> object.

```python
conn = sqlite3.connect('./data/chinook.db')
```

Once we have a <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">Connection</b> object, we can then create a <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">Cursor</b> object. Cursors allow us to execute SQL queries against a database:

```python
cur = conn.cursor()
```

The <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">Cursor</b> has created a method <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">execute</b>, which will receive SQL parameters to run against the database.
The code below will fetch the first <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">5</b> rows from the <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">employees</b> table:

```python
cur.execute('SELECT * FROM employees LIMIT 5;')
```

You may have noticed that  we didn't assign the result of the above query to a variable. THis is because we need to run another command to actually fetch the results.
We can use the <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">fetchall</b> method to fetch all of the results of a query.

```python
results = cur.fetchall()
```

```python
results
```

As you can see, the results are returned as a list of tuples. Each tuple corresponds to a row in the database that we accessed. Dealing with data this way is paintful.
We need to manually add columns headers, and manually parse the data. Luckily, the pandas library has an easier way, which we'll look at in the next section

```python
df = pd.DataFrame(results)
```

```python
df.head()
```

Before we move on, it's good practice to close <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">Connection</b> objects and <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">Cursor</b> objects that are open. This prevent the SQLite database from being locked. When a SQLite database is locked, you may be unable to update the database, and may get errors. We can colose the Cursor and the Connection like this:

```python
cur.close()
conn.close()
```

<h2>The <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">read_excel</b> method</h2> <br>

We'll begin with the <strong>read_excel</strong> method, that let us read Excel files into a <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">DataFrame</b>. <br>

This method supports both XLS and XLSX files extensions from a local filesystem or URL and has a broad set of parameters to configure how the data will be read and parsed. These parameters are very similar to the parameteres we saw on previous lectures where we introduced the <b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">read_csv</b> method. The most common parameters are as follows:

<ul style="disc">
    <li><b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">filepath</b>: Path of the file to be read.</li>
    <li><b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">sheet_name</b>: Strings are used for sheet names. Integers are used in zero-indexed sheet positions. Lists of string/integers are used to request multiple sheets. Specify None to get all sheets.</li>
    <li><b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">header</b>: Index of the row containing the names of the columns (None if none).</li>
    <li><b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">index_col</b>: Index of the column or sequence of indexes that should be used as index of rows of the data.</li>
    <li><b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">names</b>: Sequence containing the names of the columns (used together with header = None)</li>
    <li><b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">Skiprows</b>:  Number of rows or sequence of row indexes to ignore in the load</li>    
    <li><b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">na_values</b>: Sequence of values that, if found in the file, shoulb be treated as NaN</li>
    <li><b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">dtype:</b>: Dictionary in which the keys will be column names and the values will be types of NumPy to which their content must be converted</li>
    <li><b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">parse_dates</b>: Flag that indicates if Python should try to parse data with a format similar to dates as dates. You can enter a list of column names that must be joined for the parsing as a date.</li>    
    <li><b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">date_parser</b>: Function to use to try to parse dates.</li>
    <li><b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">nrows</b>:  Number of rows to read from the beginning of the file.</li>
    <li><b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">skip_footer</b>: Number of rows to read from the beginning of the file</li>
    <li><b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">squeeze</b>: Flag that indicates that if the data read only contains one column the result is a Series instead of a DataFrame</li>    
    <li><b style="font-family:georgia,garamond,serif; font-size:16px;font-style:italic;background-color:Yellow">thousands</b>:Character to use to detect the thousands separator</li>
</ul>

<h3>Dictionaries</h3>
Dictionaries are map-like collections that store values under a user-defined key. They key must be an immutable object; we usually employ strings for keys. Dictionaries are mutable, and more importantly, <strong>unordered</strong>

Access is by key, also using square brackets:
